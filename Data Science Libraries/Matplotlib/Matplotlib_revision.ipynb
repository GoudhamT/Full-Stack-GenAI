```python
x = [1,2,3,4,5]
y = [10,6,8,12,6]
```

# Line Plot

```python
plt.plot(x,y) # here x & y are co-ordinates

# Beautify code
plt.xlabel("X values")
plt.ylabel("Y label")
plt.title("My first plot")


plt.show()
```

# Bar plot

```python
plt.bar(x,y,edgecolor='Red') # here X - category and Y - frequency
plt.show()
```

```python
# Scatter plot
```

```python
plt.scatter(x,y ,marker='*') # marker is used to point in shapes

# beautify
plt.grid(False)
plt.show()
```

```python
titanic.describe()
titanic['Sex'].value_counts()
```

```python
### Age Distribution
plt.subplot(2,2,1)
plt.hist(titanic['Age'],bins=10,edgecolor='Black',label='Age')
plt.xlabel('Age')
plt.ylabel("Frequency")
plt.title("Age Distribution",color='crimson')
plt.xlim(0,60)
plt.ylim(0,200)
plt.xticks(np.arange(0,100,10))
# plt.yticks([0,80,180],['Min','Median','Max'])
plt.legend(loc='upper right')
plt.tight_layout()

### Plot using random x,y values
x = [1,2,3,4]
y = [10,4,6,12]
plt.subplot(2,2,2)
plt.plot(x,y,label="Plotting",color='#800020')
plt.xlabel('X value')
plt.ylabel('Y label')
plt.legend()

# Gender distribution
plt.subplot(2,2,4)
sex_val = titanic['Sex'].value_counts()
plt.bar(sex_val.index,sex_val.values,color=['#CD5C5C','pink'])
plt.xlabel('Sex')
plt.ylabel('Total counts')


plt.show()
```
